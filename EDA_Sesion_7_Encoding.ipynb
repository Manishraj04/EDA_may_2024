**fit and transform**

- fit mens develop the logic
- transform mean process the logic
- if we observe in map method first we develop dictionary logic
- we apply that the dict to columns
- Developing the dict is kind of fit
- apply the dict to columns means we r transforming the data cat to num
- whenever we r transforming or changing the data then use **fit and transform**
- if we r not changing the data then only fit

**np.where**

- using np.where also we can change the Categorical Data to Numerical
- But np.where is only for binary conditon
- np.where is same as if-else
- if condition is True : will apply True value
- if condition is False : will apply False value
- so np.where use for only binary labels
- A columns should have only two unique labels

```python
path=r'C:\Users\manish\Datafile\Visadataset.csv'
visa_df=pd.read_csv(path)
visa_df.head()
cat_columns=visa_df.select_dtypes(include='object').columns
```

```python
visa_df['case_status']
```

```python
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
visa_df
```
