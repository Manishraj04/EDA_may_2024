$take$

- first argument: indices
- indices will take some values in array: in a list
- it tries to extract the data
- That value we provide represents columns or row
- Axis=0 represent rows
- Axis=1 represent columns
- bydefault  axis mentioned as 0

```python
visa_df.take([100,200,300])
```
